### Analyze model performance with iter

```python
dict_results_model = {}
```

```python
path_model = os.path.join(path_main_folder, 'out', 'iterate_LR_2')
name_model = path_model.split('_')[1]
```

```python
nb_iter = 10

iter_results = pd.DataFrame(columns=['iteration','nb_ae','test_mse','test_mae', 'test_mape'])

for idx in range(nb_iter):
    
    path_folder = os.path.join(path_model,'iter_{}'.format(idx))
    results = pd.read_csv(os.path.join(path_folder,'cv_results.csv'), sep=";")
    iter_result = results[['test_mse','test_mae', 'test_mape']].mean(axis=0)
    iter_result['iteration'] = idx
    
    ael_df = pd.read_csv(os.path.join(path_folder,'ael.csv'), sep=";")
    iter_result['nb_ae'] = len(ael_df)
    
    iter_results = iter_results.append(iter_result, ignore_index=True)
```

```python
dict_results_model[name_model] = iter_results
```

```python
# model error and discovery rate with iterations
key = 'test_mape'
path_save = os.path.join(path_main_folder, 'out', 'figures')
name_fig = '{}_{}_vs_iter{}'.format(name_model, key.split('_')[1], nb_iter)

fig, ax1 = plt.subplots()

ax1.plot(iter_results['iteration'], iter_results[key],'b-o')
ax1.set_xlabel('iteration')
ax1.set_ylabel(key.split('_')[1], color = 'b')
ax1.set_title('{}: evolution of {} with the number of iteration'.format(name_model,key.split('_')[1]))
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(iter_results['iteration'], iter_results['nb_ae'],'r-o')
ax2.set_ylabel('number of different atypical events', color = 'r')
ax2.tick_params('y', colors='r')
```

```python
fig.savefig(os.path.join(path_save, name_fig + '.png'))
fig.savefig(os.path.join(path_save, name_fig + '.eps'))
```

```python
# Relative improvement with number of iterations
for key, iter_results in dict_results_model.items():
    iter_results['pct_change_mape'] = iter_results['test_mape'].pct_change()
```

```python
fig, ax = plt.subplots()

for key, iter_results in dict_results_model.items():
    ax.plot(iter_results['iteration'], -iter_results['pct_change_mape'],'-o', label=key)

ax.set_xlabel('number of iteration')
ax.set_ylabel('relative change of mape [%]')
ax.set_title('relative improvement of mape vs number of iterations')

ax.legend()
```

```python
name_fig = 'relative_imp_mape_vs_iter'
fig.savefig(os.path.join(path_save, name_fig + '.png'))
fig.savefig(os.path.join(path_save, name_fig + '.eps'))
```

```python
# number of holiday days detected per iter

# Loading reference
path_data = os.path.join("/local/home/antorosi/Documents/Prediction/data")
with open(os.path.join(path_data, 'ae_reference_list_wwe_2013-2017' + '.pickle'), 'rb') as f:
    ael_reference= pickle.load(f)
```

```python
dict_ae_results = {}
```

```python
path_model = os.path.join(path_main_folder, 'out', 'iterate_LR_2')
name_model = path_model.split('_')[1]
```

```python
type_file = '.json'
```

```python
ae_results = pd.DataFrame(columns=['iteration', 'nb_hd_tot', 'nb_other_tot', 'nb_tot', 
                                   'nb_hd', 'nb_other', 
                                   'nb_hd_new', 'nb_other_new', 'nb_new'])

nb_iter = 12 

ael = AtypicalEventsList()

for idx in range(nb_iter):
    
    path_folder = os.path.join(path_model,'iter_{}'.format(idx))
    
    ae_result = pd.Series()
    
    with open(os.path.join(path_folder, 'ael_iter'+type_file),'rb') as f:
        ael_iter= pickle.load(f)
        
    ael_inter = ael_reference.get_intersection(ael_iter)
    ae_result['nb_hd'] = len(ael_inter.set_atypical_events)
    ae_result['nb_other'] = len(ael_iter.set_atypical_events - ael_reference.set_atypical_events)

    ael_past = ael
    
    with open(os.path.join(path_folder, 'ael'+type_file),'rb') as f:
        ael = pickle.load(f)
        
    ael_inter = ael_reference.get_intersection(ael)
    ae_result['nb_hd_tot'] = len(ael_inter.set_atypical_events)
    ae_result['nb_tot'] = len(ael.set_atypical_events)
    ae_result['nb_other_tot'] = len(ael.set_atypical_events - ael_reference.set_atypical_events)
    
    new_ael = AtypicalEventsList()
    new_ael.set_atypical_events = ael.set_atypical_events - ael_past.set_atypical_events
    
    ae_result['nb_new'] = len(new_ael.set_atypical_events)
    ae_result['nb_hd_new'] = len(ael_reference.get_intersection(new_ael).set_atypical_events)
    ae_result['nb_other_new'] = len(new_ael.set_atypical_events - ael_reference.set_atypical_events)
    
    
    
    
    ae_result['iteration'] = idx
    
    ae_results = ae_results.append(ae_result, ignore_index=True)
```

```python
dict_ae_results[name_model] = ae_results
```

```python
width = 0.1
idx = 0
epsilon = 0.05

for name_model, ae_results in dict_ae_results.items():

    N = len(ae_results)
    ind = np.arange(N) # the x locations for the groups
    true_ind = ind+width*idx + epsilon*idx

    plt.bar(true_ind, ae_results['nb_hd_tot'], width, label='hd', color = 'blue')
    plt.bar(true_ind, ae_results['nb_other_tot'], width,
                 bottom=ae_results['nb_hd_tot'], label ='other', color = 'orange')

    plt.ylabel('number of events')
    plt.xlabel('iteration')
    plt.title('Total Atypical Events detected vs iteration')
    #plt.yticks(np.arange(0, 101, 10))
    plt.legend()
    
    total = ae_results['nb_tot']
    
    plt.text(true_ind[0], total[0]+5, name_model, rotation=90, verticalalignment='bottom')
    
    idx += 1
    
plt.legend(['hd', 'other'])
plt.xticks(ind+width+epsilon, [str(el) for el in list(ind)])

10
```

```python
width = 0.1
idx = 0
epsilon = 0.05

for name_model, ae_results in dict_ae_results.items():

    N = len(ae_results)
    ind = np.arange(N) # the x locations for the groups
    true_ind = ind+width*idx + epsilon*idx

    plt.bar(true_ind, ae_results['nb_hd_new'], width, label='hd', color = 'blue')
    plt.bar(true_ind, ae_results['nb_other_new'], width,
                 bottom=ae_results['nb_hd_new'], label ='other', color = 'orange')

    plt.ylabel('number of events')
    plt.xlabel('iteration')
    plt.title('New Atypical Events detected vs iteration')
    plt.yticks(np.arange(0, 71, 10))
    plt.legend()
    
    total = ae_results['nb_new']
    
    plt.text(true_ind[0], total[0]+5, name_model, rotation=90, verticalalignment='bottom')
    
    idx += 1
    
plt.legend(['hd', 'other'])
plt.xticks(ind+width+epsilon, [str(el) for el in list(ind)])

name_fig = 'new_ae_vs_iteration_iter{}'.format(N)
plt.savefig(os.path.join(path_save, name_fig + '.png'))
plt.savefig(os.path.join(path_save, name_fig + '.eps'))
```
